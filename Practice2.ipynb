```python
D = {"game":100, "Terra":1000}
for x in D:
    print(x, D[x])
```

```python
L = [10, 4, 5, 90, -9, 120, 203, -230, -240, -1, 999]
for j in range (len(L)):
    m = L[j]
    idx = j
    c = j
    for i in range(j, len(L)):
        if L[i]<m:
            m = L[i]
            idx = c
        c += 1
    tmp = L[j]
    L[j] = m
    L[idx] = tmp
print(L)
```

```python
def displaytypes(a, b, c, d):
    """Only display types"""
    print(type(a))
    print(type(b))
    print(type(c))
    print(type(d))
```

```python
displaytypes("a", 9, 9.4, {print:89})
```

```python
def checkargs(a,b,c):
    """Check if all arguments are of the same type"""
    if isinstance(a, int) and isinstance(b, int) and isinstance(c, int):
        print((a**b)/c)
        return True
    else:
        return False
```

```python
checkargs(5,3,1)
```
